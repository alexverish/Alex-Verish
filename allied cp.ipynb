```python
# import necessary Python libraries
from bs4 import BeautifulSoup
import csv, requests
import pandas as pd
import re as re
from selenium import webdriver
import time
from selenium.webdriver.support.select import Select
```

```python
name_list = []
address_list = []
city_list = []
state_list = []
zip_list = []
country_list = []
phone_list = []
url_list = []
```

```python
# call Selenium by providing path to Google Chrome webdriver (use Chrome version 73)

driver = webdriver.Chrome('C:/Users/veri8224/Documents/chromedriver')

# provide webdriver with URL link to website

driver.get('https://www.alliedcp.com/sales/distributor-locator/wyoming/')

soup = BeautifulSoup(driver.page_source,'lxml')
```

```python
listings = soup.find_all('div',{'class':'distributors_location'})

for i in range (len(listings)):
    
    name_list.append(listings[i].find('h3').text)
    
    address_list.append(listings[i].find('span').text)
    
    city_list.append(listings[i].find_all('span')[1].text.split(',')[0])
    
    state_list.append(listings[i].find_all('span')[1].text.split(',')[1].strip())
    
    zip_list.append(listings[i].find_all('span')[1].text.split(',')[2].strip())
    
    country_list.append('United States')
    
    try:
        phone_list.append(listings[i].find_all('a')[-1].text)
    except Exception:
        phone_list.append('')
    
    try:
        url_list.append(str(listings[i].find_all('a')[-2]).split('"')[1])
    except Exception:
        url_list.append('')
```

```python
df = pd.DataFrame({"name": name_list, "address": address_list, "city": city_list, "state": state_list, "zip": zip_list, "country": country_list, "phone": phone_list, "url":url_list})
```
